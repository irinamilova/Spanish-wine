## Just a picture

```python
text = df["wine"]
text = text.to_string()
tokens = text.split(" ")
word_counts = Counter(tokens)
fog_machine = WordCloud()
fog_machine.generate(text)
fog_machine.to_image()

fog_machine = WordCloud(width=800,
                        height=600,
                        min_font_size=14,
                        background_color="#333333",
                        colormap="spring")

fog_machine.generate_from_frequencies(word_counts)
fog_machine.to_image()
image = Image.open("Word_wine.png")
image
mask = np.array(image)

mask[0][0]
fog_machine = WordCloud(mask=mask,
                        max_words=7000,
                        colormap="Reds")
fog_machine.generate_from_frequencies(word_counts)
fog_machine.to_image()
```

# A hypothesis check

#### Hypothesis: Red wine with a rating of more than 4.5 produced in some regions is worth more than wine produced in other regions.

1. Отсортировать только красное вино.
2. Количество отзывов больше 50
3. Отсортировать рейтинг на 4.5 +
4. region - rating
5. Регион

```python
print(df.type.unique())
```

```python
dfr1 = df[df.type == 'Red']
dfr2 = df[df.type == 'Toro Red']
dfr3 = df[df.type == 'Ribera Del Duero Red']
dfr4 = df[df.type == 'Priorat Red']
dfr5 = df[df.type == 'Rioja Red']
dfr6 = df[df.type == 'Montsant Red']
```

```python
frames = [dfr1, dfr2, dfr3, dfr4, dfr5, dfr6]
dfrr = pd.concat(frames)
```

```python
dfrr =  dfrr[dfrr.num_reviews >= 50]
```

```python
dfrr =  dfrr[dfrr.rating >= 4.5]
```

```python
dfrr = dfrr.round(decimals = {'price': -2})
```

```python
plt.figure(figsize=(15,20))
sns.scatterplot(data = dfrr, y='region', x='year', color='black',  hue ='rating', palette="rocket", size = 'price', sizes = (20, 200))
plt.grid(axis = 'y')
plt.xticks(rotation=90);
```
